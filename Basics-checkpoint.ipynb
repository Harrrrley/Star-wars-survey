```python
import pandas as pd
import numpy as np
star_wars = pd.read_csv("star_wars.csv",encoding="ISO-8859-1")
```

```python
star_wars = star_wars[pd.notnull(star_wars["RespondentID"])]
```

```python
star_wars.head(10)
```

```python
star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].value_counts(dropna=False)
```

```python
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].value_counts(dropna=False)
```

```python
yes_no = {
    "Yes": True,
    "No": False
}
star_wars["Have you seen any of the 6 films in the Star Wars franchise?"] = \
star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].map(yes_no)
```

```python
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] = \
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].map(yes_no)
```

```python
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].value_counts(dropna=False)
```

```python
print(star_wars["Which of the following Star Wars films have you seen? Please select all that apply."].value_counts(dropna=False))
print(star_wars["Unnamed: 4"].value_counts(dropna=False))
print(star_wars["Unnamed: 5"].value_counts(dropna=False))
print(star_wars["Unnamed: 6"].value_counts(dropna=False))
print(star_wars["Unnamed: 7"].value_counts(dropna=False))
print(star_wars["Unnamed: 8"].value_counts(dropna=False))
```

```python
sw_list = ["Star Wars: Episode I  The Phantom Menace","Star Wars: Episode II  Attack of the Clones",
          "Star Wars: Episode III  Revenge of the Sith","Star Wars: Episode IV  A New Hope",
          "Star Wars: Episode V The Empire Strikes Back","Star Wars: Episode VI Return of the Jedi"]
column_name_list = star_wars.columns[3:9]
for i in range(6):
    value_transfer = {sw_list[i]:True,
                     np.NaN:False}
    star_wars[column_name_list[i]] = star_wars[column_name_list[i]].map(value_transfer)
    star_wars = star_wars.rename(columns={column_name_list[i]:"seen_"+ str(i+1)})
```

```python
star_wars.head(10)
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
name_transf = {}
for i in range(6):
    name_transf[star_wars.columns[9+i]] = "ranking_" + str(i+1)
star_wars = star_wars.rename(columns=name_transf)
```

```python
rankings = star_wars.iloc[:,9:15].mean()
```

```python
%matplotlib inline
rankings.plot.bar()
```
